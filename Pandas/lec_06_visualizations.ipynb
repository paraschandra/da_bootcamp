### Lecture 06: Visualizations

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(r'D:\VS Code\Data Analysis\da_bootcamp\Pandas\data\Ice Cream Ratings.csv')
df = df.set_index('Date')
df
```

```python
print(plt.style.available)
plt.style.use('seaborn-v0_8-dark-palette')
```

```python
df.plot(kind='line', title='Ice cream Ratings', xlabel='Daily Rating', ylabel='Scores')
```

```python
df.plot(kind='bar', stacked=True, title='Ice cream Ratings', xlabel='Daily Rating', ylabel='Scores')
```

```python
df['Flavor Rating'].plot(kind='bar', stacked=True)
```

```python
df.plot.barh(stacked = True)
```

```python
df.plot.scatter(x='Texture Rating', y='Overall Rating', s=500, c='green')
```

```python
df.plot.hist(bins=10)
```

```python
df.boxplot()
```

```python
df.plot.area(figsize=(10, 5))
```

```python
df.plot.pie(y='Flavor Rating')
```
